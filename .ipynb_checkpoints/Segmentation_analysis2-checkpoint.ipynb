## Segmentation Image 분석

```python
import tensorflow as tf
import numpy as np
import gzip
import keras
import matplotlib.pyplot as plt
import time
import random
from sklearn.gaussian_process.kernels import PairwiseKernel
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.gaussian_process import GaussianProcessRegressor
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop, Adam, Nadam
```

MNIST 데이터셋 불러오기

```python
np.set_printoptions(suppress=True)
np.set_printoptions(precision=3)

# read data set
f = gzip.open('./data/train-images-idx3-ubyte.gz', 'rb')
x_train = np.array(np.frombuffer(f.read(), np.uint8, offset=16))

f = gzip.open('./data/train-labels-idx1-ubyte.gz', 'rb')
y_train = np.array(np.frombuffer(f.read(), np.uint8, offset=8))

f = gzip.open('./data/t10k-images-idx3-ubyte.gz', 'rb')
x_test = np.array(np.frombuffer(f.read(), np.uint8, offset=16))

f = gzip.open('./data/t10k-labels-idx1-ubyte.gz', 'rb')
y_test = np.array(np.frombuffer(f.read(), np.uint8, offset=8))

# data preprocessing
x_train = np.reshape(x_train, (-1, 784))
x_train = x_train / 255
x_train = x_train.astype(np.float32)
x_test = np.reshape(x_test, (-1, 784))
x_test = x_test / 255
x_test = x_test.astype(np.float32)

# get each training images
idx_train = [0 for i in range(10)]
idx_test = [0 for i in range(10)]
for i in range(10):
    idx_train[i] = (y_train == i)
    idx_test[i] = (y_test == i)
      
x_train_38 = x_train[idx_train[3]+idx_train[8]]
x_test_38 = x_test[idx_test[3]+idx_test[8]] 

# y_train_38 = y_train[idx_train[3]+idx_train[8]][0:1000]
# print(np.count_nonzero(y_train_38[y_train_38==3]))

# one-hot encoding
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)


y_train_38 = y_train[idx_train[3]+idx_train[8]]
y_test_38 = y_test[idx_test[3]+idx_test[8]] 
```

```python
# K-Means clustering to find centers
start = time.time()
kmeans_model = KMeans(2)
kmeans_model.fit(x_train[idx_train[8]+idx_train[3]])
_38=np.reshape(kmeans_model.cluster_centers_,(-1,784))
print("Cluster Production time(sec) :", time.time() - start)
# x_model = KMeans(100)
# x_model.fit(x_train)
# x = x_model.predict(x_model.cluster_centers_)
# x = keras.utils.to_categorical(x,100)

x = np.zeros((100,100))
for i in range(100):
    x[i][i] = 1
```

```python
fig = plt.figure()

for i in range(2):
  subplot = fig.add_subplot(2,1, i+1)
  subplot.set_xticks([])
  subplot.set_yticks([])
  subplot.imshow(np.reshape(_38[i], (28, 28)), cmap='gray')
```

```python
row = [random.randrange(0,22) for i in range(50)]
col = [random.randrange(0,22) for i in range(50)]
train_label=[]
train_image =[]
test_image=[]
test_label=[]

centers=[]

for cluster in _38:
    for i in range(50):
        for j in range(0,row[i]*28):
            centers.extend([0.])

        for r in range(row[i],row[i]+7):
            centers.extend([0.]*col[i])
            centers.extend(cluster[r*28+col[i] : r*28+(col[i]+7)])
            centers.extend([0.]*(21- col[i]))

        for j in range((row[i]+7)*28,28*28):
            centers.extend([0.])

centers = np.reshape(centers,(-1,784))
```

X_Train 이미지에서 10개의 Cluster 생성후 출력  

```python
# plot cluster centers as images
imgs = centers * 255
fig = plt.figure(figsize =(5,100))
for i in range(50):
  subplot = fig.add_subplot(50,2, 2*i+1)
  subplot.imshow(np.reshape(centers[i], (28, 28)), cmap='gray')
  subplot = fig.add_subplot(50,2, 2*i+2)
  subplot.imshow(np.reshape(centers[i+50], (28, 28)), cmap='gray')
```

```python
row = [random.randrange(0,15) for i in range(50)]
col = [random.randrange(0,15) for i in range(50)]
train_label=[]
train_image =[]
test_image=[]
test_label=[]

centers=[]

for cluster in _38:
    for i in range(50):
        for j in range(0,row[i]*28):
            centers.extend([0.])

        for r in range(row[i],row[i]+14):
            centers.extend([0.]*col[i])
            centers.extend(cluster[r*28+col[i] : r*28+(col[i]+14)])
            centers.extend([0.]*(14- col[i]))

        for j in range((row[i]+14)*28,28*28):
            centers.extend([0.])

centers = np.reshape(centers,(-1,784))
```

```python
# plot cluster centers as images
imgs = centers * 255
fig = plt.figure(figsize =(5,100))
for i in range(50):
  subplot = fig.add_subplot(50,2, 2*i+1)
  subplot.imshow(np.reshape(centers[i], (28, 28)), cmap='gray')
  subplot = fig.add_subplot(50,2, 2*i+2)
  subplot.imshow(np.reshape(centers[i+50], (28, 28)), cmap='gray')
```

RBF는 입력값과 prototype간의 유클리디안 거리를 계산한다.  
입력값(input vector)은 x_train, x_test가 되며 prototype(RBF Neurons) 은 centers 가 된다.  
이를 통해 출력값(output nodes)은 이미지의 예측값이 된다.  

```python
# RBF Layer
kernel = PairwiseKernel(metric='polynomial') 
rbf_model = GaussianProcessRegressor(kernel=kernel,alpha=10).fit(centers, x)
침
temp1 = rbf_model.predict(x_train_38)
temp2 = rbf_model.predict(x_test_38)
```

RBF Training 모델 설계이다.  
batch=128, epoch=100, learning rate = 0.0005 이다.  
Centers의 개수에 맞춰 input_shape = 10 이 된다.  

```python
# Perceptron Layers
batch_size = 128
epochs = 100

model = Sequential()
model.add(Dense(10, activation='softmax', input_shape=(100,)))

model.summary()
nadam=keras.optimizers.Nadam(lr=0.0005)
model.compile(loss='categorical_crossentropy',
              optimizer=nadam,
              metrics=['accuracy'])
```

RBF 네트워크를 학습하는 과정이다. Sample이 입력되면  
RBF layer의 각 cluster center와 Radial Basis Function 값을 계산한다.  
이것이 해당 neuron의 출력값이 되며, training을 통해 weight가 결정된다.  

```python
# Train model
history = model.fit(temp1, y_train_38,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(temp2, y_test_38))
```

K-means Clustering 을 통해 MNIST training set에서 10개의 클러스터를 생성한다.  
이를 RBF Network 에서 Prototype 으로 사용하고 training sample들을 이용해 뉴런의 weight를 학습시킨다.  
이후 MNIST test sample 을 예측시킨 결과 Accuracy = 77.89% 라는 결과가 나온다.

```python
# Evaluate model
score = model.evaluate(temp2, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
